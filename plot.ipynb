```python
import matplotlib.pyplot as plt
import numpy as np

# 与えられたデータ点
data_points = [
    (1968, 2.807),
    (1978, 2.795),
    (1979, 2.779),
    (1981, 2.522),
    (1981, 2.517),
    (1981, 2.496),
    (1986, 2.479),
    (1990, 2.3755),
    (2010, 2.3737),
    (2012, 2.3729),
    (2014, 2.3728639),
    (2020, 2.3728596),
    (2022, 2.371866),
    (2024, 2.371552),
    (2025, 2.371339),
]

# データ点の分解
x_vals, y_vals = zip(*data_points)

# 直線の両端の点
x1, y1 = 1990, 2.3755
x2, y2 = 2025, 2.371339

# 傾きと切片を計算
m = (y2 - y1) / (x2 - x1)
b = y1 - m * x1

# y = 2 となるときの x を計算
x_end = (2 - b) / m

# x=2025 から y=2 までの範囲で直線を生成
x_line_segment = np.linspace(2025, x_end, 500)
y_line_segment = m * x_line_segment + b

# y=2の水平線のためのx座標を生成
x_horizontal = np.linspace(1968, x_end, 500)

# プロット
plt.figure(figsize=(10, 6))
plt.plot(x_line_segment, y_line_segment, color='red', linestyle='--')  # 直線部分
plt.plot(x_horizontal, [2] * len(x_horizontal), color='green', linestyle='-')  # y=2の水平線
plt.scatter(x_vals, y_vals, color='blue', s=20)  # データ点（サイズを20に設定）

# x軸のメモリを設定
current_ticks = plt.xticks()[0]
filtered_ticks = [tick for tick in current_ticks if tick < 2500 or tick > 5000]
plt.xticks(filtered_ticks + [5147])

# x軸の範囲を設定
plt.xlim(1968, 5200)

plt.tight_layout()

# プロットをPNGファイルとして保存
plt.savefig('plot.png', dpi=300, bbox_inches='tight')

plt.show()
```
